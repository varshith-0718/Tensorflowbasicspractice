# **Creating random tensors**
Random tensors are tensors of some arbitary size which contain random numbers.

```python
 #create two random tensors
 r1 = tf.random.Generator.from_seed(42) #set seed for reproducibility
 r1 = r1.normal(shape=(3,2))
 r2 = tf.random.Generator.from_seed(42) #set seed for reproducibility
 r2 = r2.normal(shape=(3,2))
 r1,r2,r1==r2
 # seed will help in the retaining the values of values generated in the random generator for reusability
```

# **shuffle the order of elements in a tensor**

```python
#shuffling will help neural network learn all types of images and its different patterns which will optimize the neural network
not_shuffled = tf.constant([[10,7],[3,4],[1,5]])
not_shuffled
#shuffle the before mentioned tensor
tf.random.shuffle(not_shuffled)
#it shuffles along its first dimension
```

```python
not_shuffled.ndim
```

```python
tf.random.set_seed(41) #global level random seed
tf.random.shuffle(not_shuffled,seed=42) #operation level random seed

#this will ensure it repeats the same shuffle patterns everytime it is executed
```

# **Getting information from the tensors**


*   shape
*   rank
*   axis or dimension
*   size





```python
# create a rank 4 tensor

rank_4_tensor = tf.zeros(shape=[2,3,4,5])
rank_4_tensor
```

```python
rank_4_tensor.shape,rank_4_tensor.ndim,tf.size(rank_4_tensor)
```

```python
2*3*4*5
```

```python
#get various attributes of our tensor
print("Datatype of every element:",rank_4_tensor.dtype)
print("number of dimensions(rank):",rank_4_tensor.ndim)
print("shape of tensor:",rank_4_tensor.shape)
print("Elements along the 0 axis:", rank_4_tensor.shape[0])
print("elements along the last axis",rank_4_tensor.shape[-1])
print("total number of elements in the tensor:",tf.size(rank_4_tensor))
print("total number of elements in the tensor:",tf.size(rank_4_tensor).numpy())
```

# Indexing tensors

```python
#first element from each dimension except from the second last dimension
rank_4_tensor[:1,:1,:,:1]
```

```python
#creating a rank 2 tensor(2 dimensions)
rank_2_tensor = tf.constant([[10,7],[3,4]])
rank_2_tensor.shape,rank_2_tensor.ndim
```

```python
#get the last item of each row of the rank 2 tensor
rank_2_tensor[:,-1]
```

```python
#add in extra dimension to our rank2 tensor at the end
rank_3_tensor = rank_2_tensor[...,tf.newaxis]
rank_3_tensor
```

```python
tf.expand_dims(rank_2_tensor,axis=-1)
#expand dimensions at the last axis we use -1 to access the last one
```

```python
tf.expand_dims(rank_2_tensor,axis=1)
```

Forms of aggregation



*   Get the minimum
*   Get the maximum
*   Get the mean of a vector
*   Get the sum of a tensor



```python
import numpy as np
```

```python
E = tf.constant(tf.reshape(np.random.randint(0,100,size =50),shape = (10,5)))
F = tf.reshape(E,shape=(50,))
E,F
```

```python
tf.size(E),E.shape,E.ndim
```

```python
# Find the minimum
tf.reduce_min(E)
```

```python
#find maximum
tf.reduce_max(E)
```

```python
#find the mean
tf.reduce_mean(E)
```

```python
#find the sum
tf.reduce_sum(E)
```

```python
#Find the variance of our tensor
import tensorflow_probability as tfp
tfp.stats.variance(F)
```

```python
#find the standard deviation
tf.math.reduce_std(tf.cast(F,dtype=tf.float32))
```

```python
#find the variance
tf.math.reduce_variance(tf.cast(F,dtype=tf.float32))
#we need math function for variance and standard deviation
```

# Find the positional maximum and minimum

```python
#create a new tensor for finding positional max and min
tf.random.set_seed(42)
F = tf.random.uniform(shape=[50])
F
```

```python
#Find the positional maximum
tf.argmax(F)
```

```python
#Index on our largest value position
F[np.argmax(F)]
#we can use tf.argmax() as well
```

```python
#check for equality
F[tf.argmax(F)] == tf.reduce_max(F)
```

```python
assert F[tf.argmax(F)] == tf.reduce_max(F)
```

```python
#Find the positional minimum
tf.argmin(F)
```

```python
#Finding the minimum using the positional minimum index
F[tf.argmin(F)]
```
